```python
import numpy as np
import matplotlib.image as image
import matplotlib.pyplot as plt
import scipy.misc as m
import random

im = m.face()
uimg1, uimg2, dimg1, dimg2 = np.vsplit(im, 4)
img = [i for i in range(16)]

img[0], img[1], img[2], img[3] = np.hsplit(uimg1, 4)
img[4], img[5], img[6], img[7] = np.hsplit(uimg2, 4)
img[8], img[9], img[10], img[11] = np.hsplit(dimg1, 4)
img[12], img[13], img[14], img[15] = np.hsplit(dimg2, 4)

def show():
    buds = bd.copy()
    final = np.hstack((img[buds.pop()], img[buds.pop()], img[buds.pop()], img[buds.pop()]))
    for j in range(3):
        fimg = img[buds.pop()]
        for i in range(3):
            fimg = np.hstack((fimg, img[buds.pop()]))
        final = np.vstack((final, fimg))
    plt.imshow(final)
```
